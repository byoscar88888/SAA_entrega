```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Generar datos sintéticos para la regresión
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 3 * X + np.random.randn(100, 1)  # Regresión lineal con ruido
```

```python
# Crear y entrenar el modelo de regresión lineal
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# Visualizar el ajuste de la regresión lineal
plt.scatter(X, y)
plt.plot(X, lin_reg.predict(X), color='red')
plt.title('Regresión Lineal')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

--------------------------------------------------------------------------------


Introducción a la Regresión Polinomial:

La regresión polinomial es una técnica de modelado que extiende la regresión lineal permitiendo relaciones no lineales entre las variables. A diferencia de la regresión lineal, la regresión polinomial ajusta una función polinomial a los datos, lo que proporciona más flexibilidad para capturar patrones complejos. Se utiliza cuando la relación entre las variables no es estrictamente lineal. Sin embargo, un aumento excesivo en el grado del polinomio puede conducir al sobreajuste del modelo. Es comúnmente implementada con bibliotecas como scikit-learn en Python.

```python
# Transformar los datos a características polinomiales
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)

# Crear y entrenar el modelo de regresión polinomial
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
```

```python
# Visualizar el ajuste de la regresión polinomial
plt.scatter(X, y)
plt.plot(X, poly_reg.predict(X_poly), color='green')
plt.title('Regresión Polinomial (grado 2)')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```


Este código realiza una regresión polinomial de grado 2 sobre los mismos datos que se utilizaron en la regresión lineal anterior. Aquí está la explicación detallada:

Se importan las bibliotecas necesarias, que ya están explicadas anteriormente en la explicación anterior del primer código.

Se transforman los datos originales X a características polinomiales de grado 2 utilizando PolynomialFeatures de scikit-learn. Esto significa que se generan nuevas características basadas en combinaciones polinomiales de las características originales. degree=2 especifica que se crearán características de hasta el cuadrado de las características originales, y include_bias=False indica que no se incluirá un término de sesgo (intercepto) en las características polinomiales.

Se crea una instancia de LinearRegression() llamada poly_reg.

Se entrena el modelo de regresión polinomial utilizando el método fit() con los datos transformados X_poly y y.

Se visualiza el ajuste de la regresión polinomial: se utiliza plt.scatter() para mostrar los puntos de datos reales y plt.plot() para trazar la curva de regresión polinomial ajustada. Se utiliza poly_reg.predict(X_poly) para predecir los valores de salida correspondientes a las características polinomiales transformadas X_poly. Se añaden etiquetas de título, ejes x e y para mejorar la interpretación del gráfico.

En resumen, este código genera características polinomiales de grado 2 a partir de los datos originales, ajusta un modelo de regresión lineal a estos datos transformados y visualiza el ajuste mediante un gráfico de dispersión y una curva de regresión polinomial. Esto permite modelar relaciones no lineales entre las características y la variable de respuesta.
