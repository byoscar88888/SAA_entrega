```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funciÃ³n cuadrÃ¡tica
def f(x):
    return x**2 + 5

# Definir la derivada de la funciÃ³n cuadrÃ¡tica
def df(x):
    return 2*x

# Algoritmo de descenso del gradiente
def gradient_descent(learning_rate, iterations):
    x = np.random.uniform(-10, 10)  # InicializaciÃ³n aleatoria de x
    history = [x]  # Historial de valores de x

    for _ in range(iterations):
        grad = df(x)  # Calcular el gradiente en el punto actual
        x = x - learning_rate * grad  # Actualizar el valor de x
        history.append(x)  # Agregar el nuevo valor de x al historial

    return x, history

learning_rate = 0.1
iterations = 50
optimal_x, history = gradient_descent(learning_rate, iterations)

# Imprimir el valor Ã³ptimo de x
print("Valor Ã³ptimo de x:", optimal_x)

# Graficar la funciÃ³n y el proceso de descenso del gradiente
x_values = np.linspace(-10, 10, 100)
plt.plot(x_values, f(x_values), label='FunciÃ³n cuadrÃ¡tica')
plt.scatter(history, f(np.array(history)), color='red', label='Descenso del gradiente')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Descenso del gradiente en una funciÃ³n cuadrÃ¡tica')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funciÃ³n
def f(x):
    return x**4 - 3*x**3 + 2

# Definir la derivada de la funciÃ³n
def df(x):
    return 4*x**3 - 9*x**2

# Algoritmo de descenso del gradiente
def gradient_descent(learning_rate, iterations):
    x = np.random.uniform(-2, 3)  # InicializaciÃ³n aleatoria de x
    history = [x]  # Historial de valores de x

    for _ in range(iterations):
        grad = df(x)  # Calcular el gradiente en el punto actual
        x = x - learning_rate * grad  # Actualizar el valor de x
        history.append(x)  # Agregar el nuevo valor de x al historial

    return x, history

learning_rate = 0.01
iterations = 100
optimal_x, history = gradient_descent(learning_rate, iterations)

# Imprimir el valor Ã³ptimo de x
print("Valor Ã³ptimo de x:", optimal_x)

# Graficar la funciÃ³n y el proceso de descenso del gradiente
x_values = np.linspace(-2, 3, 100)
plt.plot(x_values, f(x_values), label='FunciÃ³n')
plt.scatter(history, f(np.array(history)), color='red', label='Descenso del gradiente')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Descenso del gradiente en una funciÃ³n arbitraria')
plt.legend()
plt.show()
```
